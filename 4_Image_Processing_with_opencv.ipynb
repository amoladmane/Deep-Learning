```python
import cv2
```

## Reading coloured Image

## Open CV library will read image by default as BGR - blue green red.

```python
img1= cv2.imread('/content/Cutecat.jpeg')
img1
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img1)
```

## We need to convert image into RGB - red green blue i.e Colured

## Converitng BGR to RGB

```python
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2
```

```python
img2 = cv2.cvtColor(img1,cv2.COLOR_RGB2BGR)
img2
```

```python
plt.imshow(img2)
```

## Seprate RGB channels

```python
R,G,B=cv2.split(img2)
```

```python
R
```

```python
G
```

```python
plt.imshow(R,cmap='Blues')
```
