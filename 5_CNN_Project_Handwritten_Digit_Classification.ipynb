### pre-Process the data - Scaling Xtrain data

```python
import numpy as np
```

```python
# check unique values in ytrain.
np.unique(ytrain)
```

```python
# clearly there are diffrent 10 unique values. so this is Multi Class Regression problem.
```

```python
# Scaling XTrain and Xtest data. We know pixel values is betwwen 0 -255
```

```python
# We have already xtrain and xtest
xtrain_scl = xtrain.astype("float32")/255
xtest_scl =  xtest.astype("float32")/255
```

```python
# Now Scaling of ytrain and ytest. We will use multiclass classification metod, we will use on hot encoder ,
```

Activation Fn with Loss Fn.png

```python
from keras.utils import to_categorical
```

```python
ytrain_cat = to_categorical(ytrain)
ytest_cat = to_categorical(ytest)
```

### So we have just scaled xtrain_scl, xtest_scl, ytrain_cat, ytest_cat . Scaling is done now properly

## So, next step is create Convolutional Neural Network

image.png

###  We have diffrent layer for CNN

```python
xtrain[0].shape
```

```python
# import necessary libraries
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout, Conv2D, MaxPool2D, Flatten
```

```python
model = Sequential()
```

### Now build neural network

```python
# Initialize the model
model = Sequential()

# Add input layer
model.add(Input(shape=(28, 28, 1)))
# Add Convolution layer 1
model.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu'))
# Add max pooling layer 1
model.add(MaxPool2D(pool_size=(2,2)))
# Add Convolution layer 2
model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
# Add max pooling 2
model.add(MaxPool2D(pool_size=(2,2)))
# Add Flatten Layer
model.add(Flatten())
# Add dropout
model.add(Dropout(0.5))
# Add dense layer
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

### Next is Compilation of model

```python

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### ### Next is training model Xtrain_scl ytrain_cat

```python
hist = model.fit(xtrain_scl, ytrain_cat, validation_split=0.1, epochs=20)
```

### plot the learning curve

```python
plt.figure(figsize=(10,10))
plt.plot(hist.history['loss'], label =('train loss'))
plt.plot(hist.history['val_loss'], label=('validation loss'))
plt.xlabel('epochs')
plt.ylabel('cross entropy')
plt.legend()
plt.grid()
plt.show()
```

### Evaluate the model


```python
model.evaluate(xtrain_scl, ytrain_cat)
```

```python
model.evaluate(xtest_scl, ytest_cat)
```

### predict the results for train and test

```python
yprob_train = model.predict(xtrain_scl)
yprob_test = model.predict(xtest_scl)
```

```python
yprob_train[0:5]
```

```python
ypred_train = np.argmax(yprob_train ,axis=1)
ypred_test = np.argmax(yprob_test, axis =1)
```

## Now plot the confusion maxtrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cf = confusion_matrix(ytest, ypred_test)
cfd= ConfusionMatrixDisplay(cf)
cfd.plot()
```

## It is a good model we can say

```python
## Print the classification report
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```
